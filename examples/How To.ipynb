#### What kind of chords are currently availiable?

```python
import tempfile
import time
import IPython
from sox_chords.music import utils
from sox_chords.music import chords

# get the notes (C-B)
notes = utils.get_core_notes(4)

print("supported chords: ", map(lambda x: x.__name__,chords.chord_classes))
print("")

def print_chords(chords, name):
    print("%s chords:%s" % (name, map(lambda x: x.__name__, chords)))
    chords = [chord(note) for note in notes for chord in chords]
    class_names = [c.get_name() for c in chords]
    print("%s:%s\n" % (name, class_names))
    
# print all triads ['MajorTriad', 'MinorTriad', 'AugmentedTriad', 'DiminishedTriad']
print_chords(chords.triad.chords, "Triads")

# print all seventh ['MajorSeventh', 'MinorSeventh', 'DominantSeventh', 'MinorMajorSeventh', 'AugmentedSeventh', 'AugmentedMajorSeventh', 'DiminishedSeventh', 'HalfDiminishedSeventh']
print_chords(chords.seventh.chords, "Seventh")

print_chords(chords.sus.chords, "Sus")
```

#### How to generate a new chord?
- lets generate a Cm7 chord and play it

```python
# get the base note from 4th octave
note = chords.get_note("C", octave=4)

# build the chord
chord = chords.seventh.MajorSeventh(note)

# see what kind of notes the chord consists of
print(chord.notes)

# generate that chord using sox and putting it into a temp directory
# the chord is 1 second long
temp_path = tempfile.mkdtemp()
output_path = chord.generate(output_dir=temp_path, duration=1.0)

# play the chord
IPython.display.display(IPython.display.Audio(output_path, autoplay=True))

# when run all (wait a little here)
time.sleep(2)
```

#### Let's generate all guitar chords from E2 to E5

```python
from sox_chords.music.spezial_chords import guitar_coords

output_dir = tempfile.mkdtemp()

play = False

gc = guitar_coords([chords.triad.MajorTriad(chords.get_note("C"))])

for chord in gc:
    if play:
        output_path = chord.generate(output_dir)
        IPython.display.display(IPython.display.Audio(output_path, autoplay=True))
        time.sleep(1.5)
```

#### Lets get fancy and play some scales

```python
from sox_chords.music.scale import scales
from sox_chords.util.logger import logger

import logging
logger.setLevel(logging.DEBUG)
temp_path = tempfile.mkdtemp()


note = chords.get_note('C')
for scale_class in scales:
    scale = scale_class(note)
    print(scale_class.__name__, scale.notes)
    scale_output = scale.generate(output_dir=temp_path)
    IPython.display.display(IPython.display.Audio(scale_output))
```
